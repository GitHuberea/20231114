```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# 函數從網址加載圖片
def load_image(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content))

# 定義變換
transforms_dict = {
    "Original Image": None,
    "Resized": transforms.Resize((2048,3840)),
    "Center Cropped": transforms.CenterCrop(400),
    "Color Jittered": transforms.ColorJitter(brightness=5, contrast=5, saturation=5),
    "Horizontally Flipped": transforms.RandomHorizontalFlip(p=0.6),
    "Randomly Rotated": transforms.RandomRotation(120)
}

# 加載圖片
img_url = 'https://cdn.pixabay.com/photo/2019/06/23/04/45/cat-4292799_1280.jpg'
img = load_image(img_url)

# 準備繪圖
fig, ax = plt.subplots(1, 6, figsize=(15, 5))  # 1 row，6 column

# 繪製和顯示變換後的圖像
for i, (title, transform) in enumerate(transforms_dict.items()):
    transformed_img = img
    if transform:
        transformed_img = transform(transformed_img)
    ax[i].imshow(transformed_img)
    ax[i].set_title(title)
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# 函數從網址加載圖片
def load_image(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content))

# 定義變換
transforms_dict = {
    "Original Image": None,
    "Resized": transforms.Resize((540,1080)),
    "Center Cropped": transforms.CenterCrop(392),
    "Color Jittered": transforms.ColorJitter(brightness=1.3, contrast=1.7, saturation=2.2),
    "Horizontally Flipped": transforms.RandomHorizontalFlip(p=0.75),
    "Randomly Rotated": transforms.RandomRotation(315)
}

# 加載圖片
img_url = 'https://cdn.pixabay.com/photo/2020/08/23/02/59/capybara-5509760_1280.jpg'
img = load_image(img_url)

# 準備繪圖
fig, ax = plt.subplots(1, 6, figsize=(15, 5))  # 1 row，6 column

# 繪製和顯示變換後的圖像
for i, (title, transform) in enumerate(transforms_dict.items()):
    transformed_img = img
    if transform:
        transformed_img = transform(transformed_img)
    ax[i].imshow(transformed_img, cmap='gray')
    ax[i].set_title(title)
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```
